```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set()
```

```python
with open('languages.json', 'r', encoding='utf-8') as f:
    languages = json.load(f)

languages = dict(sorted(languages.items(), key=lambda item: item[1]))
languages = dict(reversed(list(languages.items())))

languages["nan"] = languages.pop("")

plt.figure(figsize=(20,10))
ax = sns.barplot(x=list(languages.keys())[:5], y=list(languages.values())[:5])
for i in ax.containers:
    ax.bar_label(i,)
ax.set_title('Language distribution')
ax.set_xlabel('Language')
ax.set_ylabel('Number of notebooks')
```
