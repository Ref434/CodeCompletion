# 2 在matplotlib中设置字体的常用方法

## 2.1 在matplotlib中设置全局字体

- **使用系统自带字体**

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
# for font in font_manager.fontManager.ttflist:
#     # 查看字体名以及对应的字体文件名
#     print(font.name, '-', font.fname)
```

```python
for font in font_manager.fontManager.ttflist:
    if font.fname.split('\\')[-1] == 'STHUPO.TTF':
        # 查看字体名以及对应的字体文件名
        print(font.name, '-', font.fname)
```

```python
plt.rcParams['font.sans-serif'] = 'STHupo'
plt.text(0.5, 0.5, '测试', ha='center', fontsize=50);
```

- **使用任意的字体**

```python
# 临时注册字体
font_manager.fontManager.addfont('fonts/MaShanZheng-Regular.ttf')

plt.rcParams['font.sans-serif'] = 'Ma Shan Zheng'
plt.text(0.5, 0.5, '测试', ha='center', fontsize=50);
```

## 2.2 在matplotlib中设置局部字体

```python
plt.text(0.5, 0.5, '测试', fontsize=50, ha='center',
         fontproperties=font_manager.FontProperties(fname='fonts/LiuJianMaoCao-Regular.ttf'));
```

```python
fig, ax = plt.subplots(figsize=(5, 1))

ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['刻度1', '刻度2', '刻度3'],
                   fontdict={
                       'fontproperties': font_manager.FontProperties(fname="fonts/ZCOOLKuaiLe-Regular.ttf")
                   },
                   fontsize=30);
```
