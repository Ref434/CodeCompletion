```python
import geopandas as gpd
from sqlalchemy import create_engine

chongqing = gpd.read_file('重庆市.geojson')
chongqing.plot()
```

## 2.1 向PostGIS写入数据

```python
# 创建engine
engine = create_engine('postgresql://postgres:CUDLCUDL@localhost:5432/demo')

# 向PostGIS推送数数据表
chongqing.to_postgis(name="chongqing", con=engine, if_exists='replace')
```

## 2.2 从PostGIS读入数据

```python
# 编写取数SQL语句
sql = 'SELECT * FROM chongqing'
chongqing_ = gpd.read_postgis(sql, con=engine, geom_col='geometry')
chongqing_.plot()
```
